#### **Guión para el vídeo del proyecto Nomadismo Digital**

##### **Introducción** <mark>[Miembro 1 del equipo]</mark>

***Objetivo*** 

"Hoy presentamos un análisis exploratorio de datos sobre el nomadismo digital y el coste de vida global. El objetivo: identificar qué características económicas diferencian a los mejores destinos para trabajadores remotos."

***Contexto***

Elegir un destino como nómada digital depende de 5 factores clave: 

- Coste de vida — vivienda, alimentación, transporte
- Infraestructura digital — velocidad y coste de internet
- Calidad de vida — seguridad, bienestar
- Capacidad económica local — relación salarios vs precios
- Accesibilidad — facilidad de visa

***Hipótesis a validar***

Planteamos 9 hipótesis (3 primarias + 6 secundarias):

| Primarias |                                                                |
|-----------|----------------------------------------------------------------|
| H1        | Los países mejor rankeados tienen menor coste de vida          |
| H2        | La relación salario/coste es más favorable en destinos nómadas |
| H3        | Mayor velocidad de internet = mayor score nómada               |


| Secundarias |                                                         |
|-------------|---------------------------------------------------------|
| H4          | El alquiler explica más variabilidad que alimentación   |
| H5          | Existen países con buena infraestructura a bajo coste   |
| H6          | El índice de felicidad correlaciona con el score nómada |
| H7          | Hay outliers estratégicos (bajo coste + alto score)     |
| H8          | Los destinos seguros tienen mayor score                 |
| H9          | La facilidad de visa correlaciona con el ranking        |

***Aplicación***

Este análisis es útil para: 

- Plataformas de relocación → recomendación de destinos
- Marketplaces nómadas → pricing y segmentación
- Gobiernos → atracción de talento remoto
- Empresas remote-first → políticas de compensación

***Creación de variables agregadas para validación de hipótesis***

**Variables relacionadas con el coste de vida**:

- nomad_housing_cost → promedio alquiler 1 habitación
- basic_basket_index → cesta básica de 12 productos
- daily_meal_cost → cappuccino + comida económica
- monthly_nomad_cost → coste mensual total del nómada

**Variables relacionadas con la capacidad económica**:
- local_purchasing_power → salario / coste nómada (>1 = favorable)
- housing_salary_ratio → % del salario destinado a vivienda

##### **Hallazgos interesantes*** <mark>[Todo el equipo]</mark>

***Factores determinante al elegir destino por los nómadas digitales***

1. Velocidad de acceso a internet. 
2. Bienestar social. 
3. Facilidad de acceso a visa. 
4. Poder adquisitivo. 

***Factores que no son relevantes y sorprenden a priori***

1. Los destinos mejor rankeados no siempre son los más baratos. 
2. La seguridad no parece relevante. 
3. La alimentación pesa más que el gasto en vivienda. 

***Ranking global de países donde existe mejor relación coste-retorno para nómadas***

1. Rumanía. 
2. Hungría. 
3. Lituania. 
4. Tailandia. 
5. Letonia. 

***Resumen ejecutivo***

1. La conectividad a internet es el factor clave al tratarse de un trabajo en remoto. 
2. Los nómadas buscan bienestar social, no sólo precios bajos.
